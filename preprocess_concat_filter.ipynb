```
import pandas as pd
```

```
import os
```

```
root_dir = '/content/drive/My Drive/data/hcmut'
```

```
os.listdir(root_dir)
```

```
list_df = []
for f in os.listdir(root_dir):
  if 'xlsx' in f:
    print("reading file:", f)
    df = pd.read_excel(os.path.join(root_dir, f))
    list_df += [df]
```

```
full_df = pd.concat(list_df)
```

```
# full_df.to_csv("full_data.csv", index=False)
full_df = pd.read_csv(os.path.join(root_dir, 'full_data.csv'))
```

```
list_col = ['NAMHOC', 'F_MAKH', 'F_MAMH', 'F_TENMHVN', 'MASV1', 'TKET']
```

```
fil_df = full_df[list_col]
```

```
gr = fil_df.groupby('MASV1')
```

```
count = 0
total = len(gr)
list_row = []

for mssv, diem in gr:
  diem_df = diem.pivot_table(values='TKET', columns='F_MAMH', aggfunc='first')
  diem_df['MASV1'] = mssv
  count += 1
  if count % 1000 == 0:
    print("%d/%d" %(count, total))
  list_row += [diem_df.iloc[0]]
```

```
for i in range(7):
  print("Path %d/%d" % (i+1, 7))
  start = i * 10000
  end = (i+1) * 10000
  end = min(end, total)
  df = pd.DataFrame(list_row[start:end])
  df.to_csv("file_%d.csv" % (i), index=False)
```
